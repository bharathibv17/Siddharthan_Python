### Matplotlib

usefull for making plots

```python
# importing matplotlib libraries
import matplotlib.pyplot as plt
```

```python
# import numpy to get data for our plot
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(y)
```

Plotting the data

```python
# sine wave
plt.plot(x,y)
plt.show()
```

```python
# Cosine wave
plt.plot(x,z)
plt.show()
```

```python
# adding title x axis and y axis label
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sign value')
plt.title('sine wave')
plt.show()
```

## parabola

```python
x = np.linspace(-5,5,50)
plt.plot(x, np.sin(x),'g-')
plt.plot(x, np.cos(x), 'r--')
plt.show()
```

## Bar Plot

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1]) # this will enclose in rectangle,0,0-->orgin, 1,1-->height and width
languages = ['English','French','Spanish','Latin','German'] #--> group of people speaking diff language
people =[100,50,150,40,70]
ax.bar(languages,people)
plt.xlabel('LANGUAGES')
plt.ylabel('NO OF PEOPLES')
plt.show()
```

## Pie Chart

Disturbution of entire data

```python
fig1=plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages = ['English','French','Spanish','Latin','German'] #--> group of people speaking diff language
people =[100,50,150,40,70]
ax.pie(people,labels=languages,autopct='%1.1f%%')
plt.show()
```

## Scatterplot

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='r')
plt.show()
```

## 3D ScatterPlot

```python
fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 20*np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()
```
