#### Manipulating a dataframe

```python
# adding a column to a dataframe
from sklearn.datasets import fetch_california_housing
```

```python
df_boston = fetch_california_housing()
```

```python
df_boston
```

' Here we have imported the dataset from the scikit learn'

```python
#pandas dataframe
df_boston = pd.DataFrame(california_housing,columns=california_housing.feature_names)
```

```python
df_boston.head()
```

```python
# adding a column to a dataframe
df_boston["price"] = california_housing.target
```

```python
df_boston.head()
```

```python
#removing a particular row
df_boston.drop(index=0,axis=0)
```

```python
#removing a particular column
df_boston.drop(columns='MedInc',axis=1)
```

```python
#locating a row using the index value
df_boston.iloc[2] # this gives second row value
```

```python
# locating a particular column value
print(df_boston.iloc[:,0]) # first column
print(df_boston.iloc[:,1]) # second column
print(df_boston.iloc[:,2]) # third column
print(df_boston.iloc[:,-1]) #last column   
```
